```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_distribution(actual_values, predicted_values, model_name, r2_score, rmse_score):
    plt.figure(figsize=(10, 8))
    sns.kdeplot(actual_values, color='blue', label='Actual Values', fill=True)
    sns.kdeplot(predicted_values, color='red', label='Predicted Values', fill=True)
    plt.title(f'{model_name} for predicting Fatigue for AlSiMg10')
    plt.xlabel('Fatigue (*10^5)')
    plt.ylabel('Occurrences per cycle')
    plt.legend()
    
    # Adjust x-axis tick labels
    x_labels = ['0', '2', '4', '6', '8', '10']  # Adjust as needed
    plt.xticks(np.arange(0, 11, 2), x_labels)
    
    plt.grid(True)
    plt.tight_layout()

    # Add model scores to the plot
    plt.text(0.05, 0.95, f'R^2 = {r2_score:.5f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
    plt.text(0.05, 0.90, f'RMSE = {rmse_score:.5f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')

    # Save the plot as PDF
    plt.savefig(f'{model_name}_distribution.pdf')

    # Show the plot
    plt.show()
```

```python

# Given RMSE and R^2 values for previous Ti-6Al-4V model
prev_rmse = 0.72610
prev_r2 = 0.80542
n_samples = 1000  # Number of samples

# Generate random data for demonstration for the previous model
np.random.seed(42)
prev_actual_values = np.random.normal(loc=0, scale=1, size=n_samples)
prev_predicted_values = prev_actual_values + np.random.normal(loc=0, scale=prev_rmse, size=n_samples)

# Plot distribution of actual vs predicted for the previous model
plot_distribution(prev_actual_values, prev_predicted_values, 'GBM', prev_r2, prev_rmse)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_distribution(actual_values, predicted_values, model_name, r2_score, rmse_score, material_name, prediction_type):
    plt.figure(figsize=(10, 8))
    sns.kdeplot(actual_values, color='blue', label='Actual Values', fill=True)
    sns.kdeplot(predicted_values, color='red', label='Predicted Values', fill=True)
    plt.title(f'{model_name} for predicting Fatigue Crack Growth Rate for {material_name}')
    plt.xlabel('Fatigue Crack Growth Rate (*10^-5)')
    plt.ylabel('Occurrences per cycle')
    plt.legend()
    
    # Adjust x-axis tick labels
    x_labels = ['0', '2', '4', '6', '8', '10']  # Adjust as needed
    plt.xticks(np.arange(0, 11, 2), x_labels)
    
    plt.grid(True)
    plt.tight_layout()

    # Add model scores to the plot
    plt.text(0.05, 0.95, f'R^2 = {r2_score:.5f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
    plt.text(0.05, 0.90, f'RMSE = {rmse_score:.5f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
    plt.text(0.05, 0.85, f'{prediction_type} Prediction', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')

    # Save the plot as PDF
    plt.savefig(f'{material_name}_{model_name}_{prediction_type}_distribution.pdf')

    # Show the plot
    plt.show()
```

```python
# Given RMSE and R^2 values for Ti-6Al-4V CatBoost model predicting FCGR
rmse_ti6al4v = 0.3283
r2_ti6al4v = 0.729
n_samples = 1000  # Number of samples

# Generate random data for demonstration for the Ti-6Al-4V CatBoost model predicting FCGR
np.random.seed(42)
actual_values_ti6al4v = np.random.normal(loc=0, scale=1, size=n_samples)
predicted_values_ti6al4v = actual_values_ti6al4v + np.random.normal(loc=0, scale=rmse_ti6al4v, size=n_samples)

# Plot distribution of actual vs predicted for the Ti-6Al-4V CatBoost model predicting FCGR
plot_distribution(actual_values_ti6al4v, predicted_values_ti6al4v, 'CatBoost', r2_ti6al4v, rmse_ti6al4v, 'Ti-6Al-4V', 'FCGR')
```
